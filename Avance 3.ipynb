```python


artistas_list = ['Ke Personajes', 'Ariana Grande', 'Linkin Park', 'Amelie lens', 'Maneskin']

cancion_data = []

for artista_nombre in tqdm.tqdm(artistas_list, desc='Obteniendo canciones de artistas:'):
    try:
        resultados = sp.search(q=f'artist:{artista_nombre}', type='album')

        for album in resultados['albums']['items']:
            nombre_album = album['name']
            fecha_lanzamiento = album['release_date']

            artistas = [artist['name'] for artist in album['artists']]
            artista_principal = artistas[0] if artistas else "Artista no disponible"

            artista_id = album['artists'][0]['id'] if album['artists'] else None

            año_lanzamiento = fecha_lanzamiento[:4]  # Obtiene solo el año

            # Verifica si el año está en el rango deseado
            if 2019 <= int(año_lanzamiento) <= 2024:
                # Obtiene las canciones del álbum
                tracks = sp.album_tracks(album['id'])
                for track in tracks['items']:
            
                    cancion_info = {
                        "ID Artista": artista_id,
                        "Artista": artista_principal,
                        "Nombre del Álbum": nombre_album,
                        "Nombre de la Canción": track['name'],
                        "Año de lanzamiento": año_lanzamiento
                    }
                    cancion_data.append(cancion_info)
    except spotipy.exceptions.SpotifyException as e:
        print(f"Error con el artista {artista_nombre}: {e}")
        auth_manager.cache_handler.clear()
        sp = spotipy.Spotify(auth_manager=auth_manager)
        continue

df_canciones = pd.DataFrame(cancion_data)
df_canciones
```
